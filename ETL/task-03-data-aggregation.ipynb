### Q: What are the top 3 cities in each state by sales revenue?

````
spark.sql(f"USE {environmentWidget}_{usernameWidget}_db")
````

````
%sql
SELECT 
  country,
  state,
  city,
  sales
FROM t1_bronze_orders;
````

````
%sql

CREATE OR REPLACE VIEW vw_top3_cities_report
AS 
with t1 AS (
  SELECT 
         city,
         state,
         ROUND(SUM(sales),0) AS sales_revenue
 FROM t1_bronze_orders
 GROUP BY 1,2 
 ORDER BY 2, 3 DESC
),
    t2 AS (
  SELECT 
      ROW_NUMBER() OVER (PARTITION BY state ORDER BY sales_revenue DESC) AS sales_rank,
      state,
      city,
      sales_revenue
  FROM t1 
), 
  t3 AS (
  SELECT * 
  FROM t2 
  WHERE sales_rank <= 3 
)

SELECT * 
FROM t3;
````
